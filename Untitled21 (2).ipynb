```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/bmw_cars_market_dataset_synthetic.csv")
x_pre = df.iloc[:]
y = df['price_usd'].values
```

```python
current_year = pd.Timestamp.now().year
x_pre['age_of_car'] = current_year - x_pre['year']
x_pre.drop(columns='year',inplace=True)
x_pre.drop(columns='car_id',inplace=True)
x_pre.drop(columns='price_usd',inplace=True)
# print(x_pre)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x_pre,y,test_size =0.2, random_state=0)
```

```python
from sklearn.impute import SimpleImputer
num_imputre1 = SimpleImputer(missing_values=np.nan,strategy='mean')
num_imputr2 = SimpleImputer(missing_values=np.nan,strategy='most_frequent')

# handilling the missing data for train

x_train[['age_of_car','owner_count','seats','doors','co2_emissions_g_km','fuel_consumption_l_per_100km','mileage_km','horsepower','engine_size']]=num_imputre1.fit_transform(x_train[['age_of_car','owner_count','seats','doors','co2_emissions_g_km','fuel_consumption_l_per_100km','mileage_km','horsepower','engine_size']])
x_train[['transmission','color','accident_history','service_history']] = num_imputr2.fit_transform(x_train[['transmission','color','accident_history','service_history']])
x_train.isnull().sum()

# handilling the missing data for test


x_test[['age_of_car','owner_count','seats','doors','co2_emissions_g_km','fuel_consumption_l_per_100km','mileage_km','horsepower','engine_size']]=num_imputre1.transform(x_test[['age_of_car','owner_count','seats','doors','co2_emissions_g_km','fuel_consumption_l_per_100km','mileage_km','horsepower','engine_size']])
x_test[['transmission','color','accident_history','service_history']] = num_imputr2.transform(x_test[['transmission','color','accident_history','service_history']])
x_test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
# lable encoding for train
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

x_train['transmission'] = le1.fit_transform(x_train['transmission'])
x_train['accident_history'] = le2.fit_transform(x_train['accident_history'])
x_train['service_history'] = le3.fit_transform(x_train['service_history'])

# lable encoding for test

x_test['transmission'] = le1.transform(x_test['transmission'])
x_test['accident_history'] = le2.transform(x_test['accident_history'])
x_test['service_history'] = le3.transform(x_test['service_history'])


# print(x_pre)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(handle_unknown='ignore'), ['model', 'fuel_type', 'drivetrain','body_type','color','country_sold'])], remainder='passthrough')
x_train = np.array(ct.fit_transform(x_train))
# print(x_pre)

x_test = np.array(ct.transform(x_test))
# print(x_pre)
```

```python
y_pred = lin_reg_2.predict(x_test)

print(np.concatenate((y_pred.reshape(-1,1), y_test.reshape(-1,1)), axis=1))
```

```python
def predict_car_price(*args):
    try:
        input_dict = {
            'model': args[0],
            'engine_size': args[1],
            'horsepower': args[2],
            'fuel_type': args[3],
            'transmission': args[4],
            'drivetrain': args[5],
            'mileage_km': args[6],
            'fuel_consumption_l_per_100km': args[7],
            'co2_emissions_g_km': args[8],
            'doors': args[9],
            'seats': args[10],
            'body_type': args[11],
            'color': args[12],
            'owner_count': args[13],
            'accident_history': args[14],
            'service_history': args[15],
            'country_sold': args[16],
            'age_of_car': 2026 - args[17]
        }

        df_input = pd.DataFrame([input_dict])

        df_input['transmission'] = le1.transform(df_input['transmission'])
        df_input['accident_history'] = le2.transform(df_input['accident_history'])
        df_input['service_history'] = le3.transform(df_input['service_history'])

        input_encoded = ct.transform(df_input)
        input_scaled = sc.transform(input_encoded)
        prediction = ann.predict(input_scaled)
        final_price = sc1.inverse_transform(prediction)

        return f"السعر المتوقع: {final_price[0][0]:,.2f} $"

    except Exception as e:
        return f"حدث خطأ في المعالجة: {str(e)}"
```

```python
models = list(df['model'].unique())
fuel_types = list(df['fuel_type'].unique())
transmissions = list(df['transmission'].unique())
drivetrains = list(df['drivetrain'].unique())
body_types = list(df['body_type'].unique())
colors = list(df['color'].unique())
accidents = list(df['accident_history'].unique())
services = list(df['service_history'].unique())
countries = list(df['country_sold'].unique())

interface = gr.Interface(
    fn=predict_car_price,
    inputs=[
        gr.Dropdown(models, label="موديل السيارة"),
        gr.Number(label="حجم المحرك (Engine Size)"),
        gr.Number(label="القوة الحصانية (Horsepower)"),
        gr.Dropdown(fuel_types, label="نوع الوقود"),
        gr.Dropdown(transmissions, label="ناقل الحركة"),
        gr.Dropdown(drivetrains, label="نظام الدفع"),
        gr.Number(label="المسافة المقطوعة (KM)"),
        gr.Number(label="استهلاك الوقود (L/100km)"),
        gr.Number(label="انبعاثات CO2"),
        gr.Slider(2, 5, step=1, label="عدد الأبواب"),
        gr.Slider(2, 8, step=1, label="عدد المقاعد"),
        gr.Dropdown(body_types, label="نوع الهيكل"),
        gr.Dropdown(colors, label="اللون"),
        gr.Number(label="عدد الملاك السابقين"),
        gr.Dropdown(accidents, label="تاريخ الحوادث"),
        gr.Dropdown(services, label="تاريخ الصيانه"),
        gr.Dropdown(countries, label="بلد البيع"),
        gr.Number(label="سنة الصنع")
    ],
    outputs="text",
    title="توقع أسعار سيارات BMW",
    description="أدخل مواصفات السيارة للحصول على تقدير للسعر باستخدام الذكاء الاصطناعي."
)

interface.launch(share=True)
```
